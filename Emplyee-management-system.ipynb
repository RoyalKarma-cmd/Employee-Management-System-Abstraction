```python
from abc import ABC, abstractmethod
```

```python
class Employee(ABC):
    def __init__(self, name, age, salary):
        self.name = name
        self.age = age
        self.salary = salary

    def show_details(self):
        print("\nName:",self.name)
        print("Age:",self.age)
        print("Salary:",self.salary)

   
    @abstractmethod
    def work(self):
        pass


class Developer(Employee):

    def __init__(self, name, age, salary, language, field):
        super().__init__(name, age, salary)
        self.language = language
        self.field = field
    
    def show_details(self):
        super().show_details()
        print("Language:",self.language)
        print("Field:",self.field)

    def work(self):
        print(f"\nWork: Writing {self.language} code")

class Tester(Employee):

    def __init__(self, name, age, salary, tool):
        super().__init__(name, age, salary)
        self.tool = tool

    def show_details(self):
        super().show_details()
        print("Tool:",self.tool)

    def work(self):
        print(f"\nWork: Testing Software using is {self.tool}")

class Manager(Employee):
    
    def __init__(self, name, age, salary, department):
        super().__init__(name, age, salary)
        self.department = department

    def show_details(self):
        super().show_details()
        print("Department:",self.department)

    def work(self):
        print(f"\nWork: Managing {self.department} Department")

dev = Developer("Jaydeep", 21, 50000, "Python", "Data Science")
test = Tester("Bhumi", 21, 25000, "Selenium")
manager = Manager("Shantanu", 25, 40000, "IT")

employees = [dev, test, manager]

for emp in employees:
    emp.show_details()
    emp.work()
    print("______________________________")
```
